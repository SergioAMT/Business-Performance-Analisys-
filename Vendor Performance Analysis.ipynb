## Correlation Insights

- **PurchasePrice** has weak correlations with **TotalSalesDollars** (-0.012) and **GrossProfit** (-0.016), suggesting that price variations do not significantly impact sales revenue or profit.
- **Strong correlation** between **total purchase quantity** and **total sales quantity** (0.999), confirming efficient inventory turnover.
- **Negative correlation** between **profit margin** & **total sales price** (-0.179) suggests that as sales price increases, margins decrease, possibly due to competitive pricing pressures.
- **StockTurnover** has weak negative correlations with both **GrossProfit** (-0.038) and **ProfitMargin** (-0.055), indicating that faster turnover does not necessarily result in higher profitability.


```python
num_items = len(df)
total_revenue = df['TotalSalesDollars'].sum()

df2 = df[['Description' , 'TotalSalesDollars']]
```

```python
df2['%Item'] = ((df2.index + 1) / num_items) * 100
df2['%Item'] = df2['%Item'].map('{:.1f}%'.format)

resultados = []
for i in range(len(df2)):
    if i == 0:
        valor = df2.loc[i, 'TotalSalesDollars'] / total_revenue
    else:
        valor = ((df2.loc[i, 'TotalSalesDollars'] / total_revenue) + resultados[i-1]) 
    resultados.append(valor)
```

```python
df2['%Revenue'] = resultados
df2['%Revenue'] = df2['%Revenue'] * 100
```

```python
ABC = []
for i in range(len(df2)):
    current_valor = df2.loc[i, '%Revenue']
    
    if current_valor <= 40:
        valor = 'A'
    elif current_valor <= 80:
        valor = 'B'
    else:
        valor = 'C'
        
    ABC.append(valor)

df2['ABC'] = ABC


df2['%Revenue'] = df2['%Revenue'].map('{:.1f}%'.format)
```

```python
df2[(df2['ABC'] == 'A') | (df2['ABC'] == 'B')]['Description']

len(df2['Description'].unique())
```

```python
# Asegurarte de que %Item y %Revenue estn como nmeros (en caso de que los hayas formateado como string con %)
df2['%Item_num'] = df2['%Item'].str.replace('%', '').astype(float)
df2['%Revenue_num'] = df2['%Revenue'].str.replace('%', '').astype(float)

# Obtener el ltimo valor de A y de B
ultimo_A = df2[df2['ABC'] == 'A'].iloc[-1]
ultimo_B = df2[df2['ABC'] == 'B'].iloc[-1]

# Crear el grfico
plt.figure(figsize=(8, 6))
plt.plot(df2['%Item_num'], df2['%Revenue_num'])

# Marcar y dibujar "L" para el ltimo A
plt.plot([ultimo_A['%Item_num'], ultimo_A['%Item_num']], [0, ultimo_A['%Revenue_num']], color='red', linestyle='--')
plt.plot([0, ultimo_A['%Item_num']], [ultimo_A['%Revenue_num'], ultimo_A['%Revenue_num']], color='red', linestyle='--')
plt.scatter(ultimo_A['%Item_num'], ultimo_A['%Revenue_num'], color='red', zorder=5, label="ltimo A")

# Marcar y dibujar "L" para el ltimo B
plt.plot([ultimo_B['%Item_num'], ultimo_B['%Item_num']], [0, ultimo_B['%Revenue_num']], color='orange', linestyle='--')
plt.plot([0, ultimo_B['%Item_num']], [ultimo_B['%Revenue_num'], ultimo_B['%Revenue_num']], color='orange', linestyle='--')
plt.scatter(ultimo_B['%Item_num'], ultimo_B['%Revenue_num'], color='orange', zorder=5, label="ltimo B")


# Etiquetas y ttulo
plt.xlabel('% Items')
plt.ylabel('% Incomes')
plt.title('Curva de Pareto: %Items vs %Revenue')
plt.grid(True)

plt.show()
```

Identify Brands that needs Promotional or Pricing adjustments which exhibit lower sales performance but higher profit margins

```python
brand_performance = df.groupby('Description').agg({
    'TotalSalesDollars' : 'sum',
    'ProfitMargin' : 'mean'
}).reset_index()
brand_performance
```

```python
low_sales_threshold = brand_performance['TotalSalesDollars'].quantile(0.15)
high_margin_threshold = brand_performance['ProfitMargin'].quantile(0.85)
```

```python
print(low_sales_threshold, high_margin_threshold)
```

```python
#Filter brands with low sales but high profit margins
target_brands = brand_performance[
    (brand_performance['TotalSalesDollars'] <= low_sales_threshold) &
    (brand_performance['ProfitMargin'] >= high_margin_threshold)
]

print("Brands with low sales but high profit margins:")
display(target_brands.sort_values('TotalSalesDollars'))
```

```python
brand_performance = brand_performance[brand_performance['TotalSalesDollars'] < 10000] #FOR BETTER VISUALIZATION
```

```python
def format_dollars(value):
    if value >= 1_000_000:
        return f"{value / 1_000_000:.2f}M"
    elif value >= 1_000:
        return f"{value / 1_000:.2f}K"
    else:
        return str(value)
```

```python
top_vendors = df.groupby('VendorName')['TotalSalesDollars'].sum().nlargest(10)
top_brands = df.groupby('Description')['TotalSalesDollars'].sum().nlargest(10)
top_vendors.apply(lambda x : format_dollars(x))
```

```python
top_brands.apply(lambda x : format_dollars(x))
```

```python
# Create subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 8))

# Plot 1: Top Vendors
top_vendors.plot(kind='barh', ax=ax1, color='skyblue')
ax1.set_title('Top 10 Vendors by Total Sales', fontsize=14, fontweight='bold')
ax1.set_xlabel('Total Sales ($)', fontsize=12)
ax1.set_ylabel('Vendor Name', fontsize=12)
ax1.grid(axis='x', alpha=0.3)

# Add value labels to vendor bars
for i, (index, value) in enumerate(top_vendors.items()):
    ax1.text(value + max(top_vendors) * 0.01, i, f'${value/1e6:.1f}M', 
             va='center', ha='left', fontweight='bold')

# Format x-axis labels as currency
ax1.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'${x/1e6:.1f}M'))

# Plot 2: Top Brands
top_brands.plot(kind='barh', ax=ax2, color='lightcoral')
ax2.set_title('Top 10 Brands by Total Sales', fontsize=14, fontweight='bold')
ax2.set_xlabel('Total Sales ($)', fontsize=12)
ax2.set_ylabel('Brand Description', fontsize=12)
ax2.grid(axis='x', alpha=0.3)

# Add value labels to brand bars
for i, (index, value) in enumerate(top_brands.items()):
    ax2.text(value * 0.98, i, f'${value/1e6:.1f}M', 
             va='center', ha='right', fontweight='light', color='white')

# Format x-axis labels as currency
ax2.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'${x/1e6:.1f}M'))

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()

```

Which vendors contribute the most to total purchase 

```python
vendor_performance = df.groupby("VendorName").agg({
    'TotalPurchaseDollars': "sum",
    'GrossProfit': "sum",
    'TotalSalesDollars': "sum"
}).reset_index()
vendor_performance
```

```python
totalPurchaseDollars = vendor_performance['TotalPurchaseDollars'].sum()

vendor_performance['PurchaseContribution%'] = (vendor_performance['TotalPurchaseDollars'] / totalPurchaseDollars)*100
vendor_performance = round(vendor_performance.sort_values('PurchaseContribution%', ascending=False),2)
```

```python
#Display top 10 vendors
top_vendors = vendor_performance.head(10)
top_vendors['TotalSalesDollars'] = top_vendors['TotalSalesDollars'].apply(format_dollars)
top_vendors['TotalPurchaseDollars'] = top_vendors['TotalPurchaseDollars'].apply(format_dollars)
top_vendors['GrossProfit'] = top_vendors['GrossProfit'].apply(format_dollars)
top_vendors
```

```python
top_vendors['PurchaseContribution%'].sum()
```

```python
top_vendors['Cumulative_contribution%'] = top_vendors['PurchaseContribution%'].cumsum()
top_vendors
```

```python
fig, ax1 = plt.subplots(figsize=(10, 6))

#bar plot for Purchase Contribution%
sns.barplot(x=top_vendors['VendorName'], y=top_vendors["PurchaseContribution%"], palette='mako', ax=ax1)

for i, value in enumerate(top_vendors['PurchaseContribution%']):
    ax1.text(i, value - 1, str(value)+'%', ha='center', fontsize=10, color='white')
    
#line plot for cumulative contribution%

ax2 = ax1.twinx()
ax2.plot(top_vendors['VendorName'],top_vendors["Cumulative_contribution%"], color='red', marker='o', linestyle='dashed', label='Cumulative %' )

ax1.set_xticklabels(top_vendors['VendorName'], rotation=90)
ax1.set_ylabel('Purchase Contribution %', color='blue')
ax2.set_ylabel('Cumulative Contribution %', color='red')
ax1.set_xlabel('Vendors')
ax1.set_title('Pareto Chart: Vendor contribution to Total Purchase')

ax2.axhline(y=100, color='gray', linestyle='dashed', alpha=0.7)
ax2.legend(loc='upper right')
    
plt.show()
```

How much capital is locked in unsold inventory per vendor, and which vendors contribute the most to it?

```python
df['UnsoldInventoryValue'] = (df['TotalPurchaseQuantity'] - df['TotalSalesQuantity']) * df['PurchasePrice']
print('Total Unsold Capital:', format_dollars(df['UnsoldInventoryValue'].sum()))
```

```python
#Aggregare Capital per Vendor
inventoryValuePerVendor = df.groupby('VendorName')['UnsoldInventoryValue'].sum().reset_index()

#Sort Vendors with the Highest Locked Capital
inventoryValuePerVendor = inventoryValuePerVendor.sort_values('UnsoldInventoryValue', ascending=False)
inventoryValuePerVendor['UnsoldInventoryValue'] = inventoryValuePerVendor['UnsoldInventoryValue'].apply(format_dollars)
inventoryValuePerVendor.head(10)
```
